```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime
from datetime import datetime
import matplotlib.dates as mdates

data = {'time': ['15:50:35', '15:50:35', '15:50:35', '15:50:35', '15:50:35', '15:50:37', '15:50:39', '15:50:37', '15:50:39', '15:50:37', '15:50:39', '15:50:37', '15:50:39', '15:50:37', '15:50:39', '15:50:41', '15:50:43', '15:50:41', '15:50:43', '15:50:41', '15:50:43', '15:50:41', '15:50:43', '15:50:41', '15:50:43', '15:50:45', '15:50:47', '15:50:49'], 'temperature': ['23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0 ', '23.0625 ', '23.0625 ', '23.0 ']}


data2 = {'time': [],
        'temperature' : []}

for x in data['temperature']:
    data2['temperature'].append(float(x))

for x in data['time']:
    data2['time'].append(mdates.date2num(datetime.strptime(x,"%H:%M:%S")))

## Converter function
#datefunc = lambda x: mdates.date2num(datetime.strptime(x,"%H:%M:%S"))


fig = plt.figure()
ax = fig.add_subplot(111)

# Configure x-ticks
ax.set_xticks(data2['time']) # Tickmark + label at every plotted point
ax.xaxis.set_major_formatter(mdates.DateFormatter("%H:%M:%S"))

ax.plot_date(data2['time'], data2['temperature'], ls='-', marker='o')
ax.set_title('title')
ax.set_ylabel('Waterlevel (m)')
ax.grid(True)

# Format the x-axis for dates (label formatting, rotation)
fig.autofmt_xdate(rotation=45)
fig.tight_layout()

fig.show()
```
